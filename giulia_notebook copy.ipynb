```python
import pandas as pd # data processing
import pandas_profiling as pp

import numpy as np # linear algebra
```

```python
df_netflix = pd.read_csv('originalDataset/netflix_titles.csv')
df_amazon = pd.read_csv('originalDataset/amazon_prime_titles.csv')
df_disney = pd.read_csv('originalDataset/disney_plus_titles.csv')
df_hulu = pd.read_csv('originalDataset/hulu_titles.csv')
df_mix = pd.read_csv('originalDataset/MoviesOnStreamingPlatforms.csv')
df_imbd = pd.read_csv('originalDataset/IMDB_dataset.csv')
```

```python
df_mix = df_mix.reset_index(drop=True)
df_mix['Title'] = df_mix['Title'].replace({'"':''}, regex=True)
df_mix.drop(columns = df_mix.columns[0], axis = 1, inplace= True)
df_mix.rename(columns = {'Prime Video':'PrimeVideo', 'Rotten Tomatoes':'RottenTomatoes'}, inplace = True)
df_mix.head()
```

```python
movie = []
for i, r in df_mix.iterrows():
    movie.append((r['Title'], r['Year']))
print(movie)
```

```python
def filtering(df):

    age = []
    rate = []
    netflix = []
    hulu = []
    prime = []
    disney = []

    new = df.copy()
    for i, r in new.iterrows():
        if (r['title'],r['release_year']) in movie:
            q = df_mix.query('Title == "'+ r['title'] +'"')
            for x in q['Age'].iteritems():
                age.append(x[1])
            for x in q['RottenTomatoes'].iteritems():
                rate.append(x[1])
            for x in q['Netflix'].iteritems():
                netflix.append(x[1])
            for x in q['Hulu'].iteritems():
                hulu.append(x[1])
            for x in q['PrimeVideo'].iteritems():
                prime.append(x[1])
            for x in q['Disney+'].iteritems():
                disney.append(x[1])
        else:
            new.drop([i], axis=0, inplace=True)

    new['rotten_rate'] = rate
    new['netflix'] = netflix
    new['hulu'] = hulu
    new['prime'] = prime
    new['disney'] = disney
    new = new.reset_index(drop=True)
    new.drop(columns = new.columns[0], axis = 1, inplace= True)

    return new

new_netflix = filtering(df_netflix)
new_amazon = filtering(df_amazon)
new_hulu = filtering(df_hulu)
new_disney = filtering(df_disney)
```

```python
new_netflix.rename(columns = {'date_added':'date_added_netflix'}, inplace = True)
new_amazon.rename(columns = {'date_added':'date_added_amazon'}, inplace = True)
new_hulu.rename(columns = {'date_added':'date_added_hulu'}, inplace = True)
new_disney.rename(columns = {'date_added':'date_added_disney'}, inplace = True)
```

```python
print(len(new_netflix), len(new_amazon), len(new_hulu), len(new_disney))
```

```python
all = pd.concat([new_netflix, new_amazon, new_hulu, new_disney],axis=0, join="outer", sort=False)

all
```

```python
result = all.copy()
result.replace(np.nan, 'null', inplace=True)
k = 0
l = 0
for i, r in result.iterrows():
    if r['netflix'] == 1 and r['hulu'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_hulu'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_hulu'] = el
        for x in q['date_added_netflix'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_netflix'] = el
    
    if r['netflix'] == 1 and r['prime'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_amazon'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_amazon'] = el
        for x in q['date_added_netflix'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_netflix'] = el

    if r['netflix'] == 1 and r['disney'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_disney'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_disney'] = el
        for x in q['date_added_netflix'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_netflix'] = el

    if r['prime'] == 1 and r['hulu'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_hulu'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_hulu'] = el
        for x in q['date_added_amazon'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_amazon'] = el
    
    if r['prime'] == 1 and r['disney'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_disney'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_disney'] = el
        for x in q['date_added_amazon'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_amazon'] = el
    
    if r['disney'] == 1 and r['hulu'] == 1:
        q = result.query('title == "'+ r['title'] +'"')
        for x in q['date_added_hulu'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_hulu'] = el
        for x in q['date_added_disney'].iteritems():
            el = x[1]
            if el != 'null':
                result.loc[i, 'date_added_disney'] = el

    if 'min' in r['rating']:
        k += 1
        if r['duration'] != 'null':
            l += 1
            result.loc[i, 'duration'] = r['rating']
            result.loc[i, 'rating'] = 'null'

    
```

```python
result.replace( 'null', np.nan, inplace=True)
result.reset_index(drop=True)
result.drop_duplicates(subset=['title', 'rating'], keep='first')

pp.ProfileReport(result, minimal = True)
```

```python
df = result.query("type == 'Movie'")
df = df.reset_index(drop=True)
df.drop(columns = df.columns[0], axis = 1, inplace= True)
df
```
